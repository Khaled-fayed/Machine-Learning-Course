# Question 2

(a) Write a short python program to plot J (t) vs. t for 100 values of t in the interval t âˆˆ [0, 5].

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([0,1.3,2.1,2.8,4.2,5.7])
y = np.array([-1,-1,-1,1,-1,1]) 
tvals = np.linspace(0,5,100)
J = []
for t in tvals:
    z = x-t
    Jt = np.sum(np.maximum(0,1-y*z))
    J.append(Jt)     
J = np.array(J)

plt.plot(tvals, J)
plt.xlabel('t') 
plt.ylabel('J(t)') 
plt.grid()
```

(b) Based on the plot, what is one value of t that minimizes J(t).

Answer: t = 4 is a minimizer.

(a) Use python to plot z vs. x and yhat vs.x when Î³ = 3 and Î± = [0,0,1,1].

```python
x = np.array([0,1,2,3])
y = np.array([1,-1,1,-1])

def plot_radial_basis(x,y,a,gamma):
    xp = np.linspace(-2,5,100)
    xpmat, xmat = np.meshgrid(xp,x)
    dist = (xpmat-xmat)**2
    z = (y*a).dot(np.exp(-gamma*dist))
    plt.plot(xp,z)
    plt.grid()
    plt.xlabel('x', fontsize = 16)
a = np.array([0,0,1,1])
plot_radial_basis(x,y,a,gamma = 3)
plt.title('gamma = 3')
```

(b) Repeat (a) with Î³ = 0.3 and Î± = [1,1,1,1].

```python
a = np.array([1,1,1,1])
plot_radial_basis(x,y,a,gamma = 0.3)
plt.title('gamma = 0.3')
```
